This notebook contains code for scraping Roblox app reviews on the Play Store using `google-play-scrapper` (https://pypi.org/project/google-play-scraper/).

```python
from google_play_scraper import reviews, Sort
import pandas as pd
from google.colab import files
```

Take reviews from Roblox in Google Playstore for all stars rating and limit it to 3052 reviews based on newest reviews in Indonesia with English language

```python
scrapreview1, continuation_token = reviews(
    "com.roblox.client",          # ID app
    lang="en",                    # defaults to ‘en’
    country="id",                 # defaults to ‘us’
    sort=Sort.NEWEST,             # defaults to Sort.MOST_RELEVANT
    count=2012,                   # jumlah ulasan yang diinginkan
    filter_score_with=1           # defaults to None (means all score)
)

app_reviews_df = pd.DataFrame(scrapreview1)
app_reviews_df.to_csv("/content/roblox_playstore1.csv", index=None, header=True)
files.download("roblox_playstore1.csv")

df_1 = pd.read_csv("/content/roblox_playstore1.csv")
df_1
```
